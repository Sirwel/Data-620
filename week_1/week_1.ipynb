```python
import networkx as nx
import matplotlib.pyplot as pl
```

```python
G = nx.Graph()
G.add_nodes_from([
    "Marleni", "Laura",
                 "Daylan", "Lewris", "Mariela", "Eugenio", "Rafael", "Miguel", "Ed", "Jonh"
                 ])
G.add_edges_from([
    ("Marleni", "Laura"),
    ("Laura", "Daylan"),
    ("Daylan", "Lewris"),
    ("Daylan", "Mariela"),
    ("Mariela", "Lewris"),
    ("Mariela", "Eugenio"),
    ("Mariela", "Rafael"),
    ("Mariela", "Miguel"),
    ("Lewris", "Miguel"),
    ("Lewris", "Ed"),
    ("Lewris", "Jonh"),
    ("Miguel", "Ed"),
    ("Miguel", "Eugenio"),
    ("Ed", "Jonh"),
    ("Eugenio", "Rafael"),
    ("Rafael", "Jonh"),
    ("Miguel", "Rafael"),
    ("Miguel", "Jonh")
])
```

```python
pos = nx.spring_layout(G, seed=1)
nx.draw(G, pos, with_labels=True,node_size=2000,font_size=12)
pl.plot()
pl.savefig("graph.png", format="png", dpi=300, bbox_inches="tight")
```
